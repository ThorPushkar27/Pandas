# Concat() and Merge() functions

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({"A":[1,2,3],"B":[4,5,6],"C":[6,7,8]})
df2 =pd.DataFrame({"A":[1,2,3],"D":["Ankur","Faltu","harsh"],"E":["Hey","Bye","Guy"]})
#During merging two data frames at least one of the values must be same.
pd.merge(df1,df2,on = "A", indicator = True) #Common part should be given using "on = " function

#Remmenber using how function.
```

```python
df1 = pd.DataFrame({"A":[1,2,3],"B":[4,5,6],"C":[6,7,8]})
df2 =pd.DataFrame({"A":[1,2,3],"B":["Ankur","Faltu","harsh"],"C":["Hey","Bye","Guy"]})
#During merging two data frames at least one of the values must be same.
pd.merge(df1,df2,left_index = True, right_index = True,suffixes= ("Name", "Id_no.")) #Common part should be given using "on = " function
```

# Concat()

```python
s1 =pd.Series([1,2,3,4,5])
s2 = pd.Series([9,8,7,6,5])
v = pd.concat([s1,s2])
v.ignore_index = True
v
```

```python
df1 = pd.DataFrame({"A":[1,2,3],"B":[4,5,6],"C":[6,7,8]})
df2 =pd.DataFrame({"A":[1,2,3],"B":["Ankur","Faltu","harsh"],"C":["Hey","Bye","Guy"]})
pd.concat([df1,df2],axis = 1)
#if the data is not equal in every list then it will give NaN values there.
#Concat doesnot use common dataframe to merge, this is why it is different from merge function.
```
