```python
x = ["Python","C","C++", "JAVA"]
y = [85,70,60,82]
z = [10,20,30,40]  
plt.title("Pushkar")
plt.xlabel("Prog. language") #we can also change the font size of the label.
plt.ylabel("Popularity in %")
c = ["Red","Green","Blue","Yellow"]
plt.bar(x,y, width = -0.6, color = c, align= "center", edgecolor = "k", linewidth = 0.8) #if we give "c", it will give cyan coloured graphs.
#align takes only 2 parameter, centre and edge, by default "center".
#we can also change the color of the edge.
#go for color codes of matplotlib.
#alpha parameter makes your graph dim.

plt.bar(x,z, align = "center", color = "c", width = 0.6, linewidth = 5, linestyle= ":")
plt.show()
```

```python
#Horizontal Bar Graph
x = ["Python","C","C++", "JAVA"]
y = [85,70,60,82]
z = [10,20,30,40]
plt.title("Pushkar")
plt.xlabel("Prog. language") #we can also change the font size of the label.
plt.ylabel("Popularity in %")

plt.barh(x,y, color = "k") #barh is used for horizontal bar graph.
plt.show()
```

# Pie Chart

```python
#syntax 
# x = []
#plt.pie(x)
#plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ["Ankur","Harsh","Himanshu","Aditya"]

ex = [0.2,0,0,0.0]
c = ["r","k","y","g"]
plt.pie(x, labels = y, explode = ex, colors = c, autopct ="%0.1f%%", shadow = True, radius = 1.0) #labels not label
#explode parameter is used to make highlight a piece out of a pie.
#function used to generate % value in pie is "autopct ="%0.1f%%".
plt.show()
```

```python
x = [10,20,30,40]
y = ["Ankur","Harsh","Himanshu","Aditya"]

ex = [0.2,0,0,0.0]
c = ["r","k","y","g"]
plt.pie(x, labels = y, explode = ex, colors = c, autopct ="%0.1f%%", shadow = True, radius = 1.0, startangle = 90) #labels not label
#explode parameter is used to make highlight a piece out of a pie.
#function used to generate % value in pie is "autopct ="%0.1f%%".
#startangle can be used to rotate the pie chart.
plt.show()
```

```python
x = [10,20,30,40]
y = ["Ankur","Harsh","Himanshu","Aditya"]

ex = [0.2,0,0,0.0]
c = ["r","k","y","g"]
plt.pie(x, labels = y, explode = ex, colors = c, autopct ="%0.1f%%", shadow = True, radius = 1.5, startangle = 90,textprops= {"fontsize": 15}, counterclock = False)

plt.title("Marks", size  = 45, color = "m")
plt.legend(loc= 1) #loc parameter gives the place where to give legend.
plt.show()
```

```python
#Dot Pie chart
plt.pie([1]) #used to help making a tube shaped pie chart if we make its color white.
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8]
y = [-2,3,-4,5,7,-9,8,6]
plt.title("Hakuna Matata")
plt.stem(x,y, linefmt= ":",  markerfmt= "ro", bottom = 2,basefmt= "g", label= "Ankur", use_line_collection= False)
#for horizontal plot we can use (orientation="horizontal")
plt.legend()
plt.show()
```

# Box Plot & Whisker Plot

```python
#x = []
#plt.boxplot(x)
#plt.show()
```

```python
import matplotlib.pyplot as lt
x =[10,20,30,40,50,60,70,120,150]
lt.boxplot(x,widths = 0.1, labels = ["Python"],
           patch_artist = True, showmeans= True, sym = "r", boxprops = dict(color ="r"), capprops= dict(color="g"))
#showmeans shows the mean of the graph shown.
# o shown on the top of the graph is called the outlier. 
lt.show()
```

```python
x =[10,20,30,40,50,60,70,120,150]
x1= [20,10,50,45,56,89,90,97,89,54]
y = [x,x1]
plt.boxplot(y, labels = ["Ankur","Harsh"])
plt.show()
```

# Step Plot

```python
#step plot is used to signal in the communication system.

#syntax
#x= []
#y = []
#plt.step(x,y)
#plt.show()
```

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]
plt.title("ANKUR")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.step(x,y, color = "r", marker = "o", ms =10, mfc = "k", label = "ANKUR")
plt.legend()
#ms = marker size
#mfc = marker face color.
plt.grid()
plt.show()
```
